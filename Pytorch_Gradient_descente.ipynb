# Méthode de gradient descente 

Le Machine learning supervisé est un type d'apprentissage automatique ou à partir de données annotés, on entraine un modèle pour appliquer ensuite ce modèle a des données non annotés.
La méthode d'optimisation de gradient descente s'applique sur une fonction f(x) pour trouver les extremums d'une fonction non dérivable.
Exemple : Pour évaluer notre modèle de ML supervisé, on va comparé les valeurs données par notre modèle aux valeurs exactes qui sont les annotations de nos données (les bonnes réponses). On cherche donc a minimiser h(x) = f(x) - y. C'est nontre fonction loss, notre fonction de perte, notre fonction d'erreur. Lorsqu'on trouve le minimum de h(x) donc on peux appliquer f(x) comme fonction de prédiction à un nouveau jeu de donnée.

```python
def f(x, y):
    return x ** 2 + y ** 2
```

```python
x=np.linspace(-6,6,30)
y=np.linspace(-6,6,30)
X, Y = np.meshgrid(x, y)
Z = f(X,Y)
```

```python
# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, rcount=50, cmap='binary')

ax.set(xlabel='Paramètre 1',
       ylabel='Paramètre 2',
       zlabel='Erreur')
ax.view_init(30,30)

plt.show()
```

## Création des deux variables

```python
x = torch.tensor(3, dtype=torch.float, requires_grad=True)
y = torch.tensor(3, dtype=torch.float, requires_grad=True)
```

Grace a pytorch on va cherche le minimum de la fonction ci-dessus avec la méthdie de gradient de descente

```python
for i in range(100):
    #on calcul le gradient de chaque variable et 
    #on applique la forme de la méthode de gradient de descente
    result = x**2 + y**2
    print("result = >>", result)
    
    result.backward() #calcule du gradient pour chaque opération faite avant
    
    print(x.grad)
    print(y.grad)
    
    # Application de la méthode de descente de gradient avec pytorch
    with torch.no_grad():
        x -= 0.1*x.grad
        y -= 0.1*y.grad
    
    #remet tous les gradient à zéro
    x.grad.zero_() #underscore automatiquement applique l'opération aux gradient
    y.grad.zero_()
    
```
